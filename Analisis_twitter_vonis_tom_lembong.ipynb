```python
# Ambil kolom yang dibutuhkan (username dan full_text)
df_selected = df_cleaned[['username', 'full_text']]

# Tampilkan 5 baris pertama dari DataFrame yang baru
df_selected
```

```python
# tokenization
import nltk
nltk.download('stopwords')
nltk.download('punkt')
import string
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Indonesian stopwords
stop_words_id = set(stopwords.words('indonesian'))

# Add custom stop words if needed
custom_stop_words = set(string.punctuation)
stop_words = stop_words_id.union(custom_stop_words)

# Normalization dictionary for informal words
informal_words = {
    "yg": "yang",
    "gak": "tidak",
    "ngisi": "mengisi",
    "udah": "sudah",
    "d": "di", # This might be too general, consider context
    "sih": "", # Remove common informal particles
    "nya": "", # Remove common informal particles
    "srg": "sering",
    "utk": "untuk",
    "byk": "banyak",
    "gk": "tidak",
    "ga": "tidak",
    "aja": "saja",
    "tp": "tapi",
    "udh": "sudah",
    "bhw": "bahwa", # Added bhw based on previous output
    "ttp": "tetap", # Added ttp based on previous output
    "tdk": "tidak",
    "tidak": "tidak"
    # Added tdk based on previous output
}


def normalize_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove mentions
    text = re.sub(r'@\w+', '', text)
    # Remove links (URLs)
    text = re.sub(r'http[s]?://\S+', '', text)
    # Tokenize
    tokens = word_tokenize(text)
    # Normalize informal words, remove punctuation and stop words
    cleaned_tokens = [informal_words.get(word, word) for word in tokens if word.isalpha() and word not in stop_words]
    # Join tokens back into a string
    return ' '.join(cleaned_tokens)

# Apply the cleaning function to the 'full_text' column
df_selected['full_text'] = df_selected['full_text'].apply(normalize_text)

# Display the cleaned DataFrame
df_selected
```

```python
# labelling
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize the VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Create sentiment lexicons (example - you can expand this)
positive_words = set(["adil", "baik", "benar", "dukung", "setuju", "apresiasi", "bijak", "tegas", ])
negative_words = set(["salah", "korupsi", "janggal", "langgar", "rugi", "kecewa", "masalah", "berat", "dipidana", "taruhan", "kriminalisasi", "keliru", "azab", "sesat", "abai", "rezim", "tidak", "aneh", "intervensi", "tidak adil"])
neutral_words = set(["vonis", "hakim", "pengadilan", "pidana", "kasus", "perkara", "sidang", "proses", "hukum"])

# Function to determine sentiment based on lexicons
def classify_sentiment_lexicon(text):
    words = set(text.split())
    pos_score = len(words.intersection(positive_words))
    neg_score = len(words.intersection(negative_words))

    if pos_score > neg_score:
        return 'Positive'
    elif neg_score > pos_score:
        return 'Negative'
    else:
        return 'Neutral'

# Apply the lexicon-based sentiment classification
df_selected['sentiment_lexicon'] = df_selected['full_text'].apply(classify_sentiment_lexicon)

# Display the DataFrame with the new sentiment column
df_selected
```

# Resampling Target

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train, y_train = smote.fit_resample(X_train, y_train)

plt.figure(figsize=(12, 6))
sns.countplot(x=y_train)
plt.title('Distribusi target untuk modeling')
plt.show()
```

# Modelling

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV

# Initialize the Naive Bayes classifier
nb_model = MultinomialNB()

# Define the hyperparameters to tune
param_grid = {'alpha': [0.1, 0.5, 1.0, 1.5, 2.0]} # Alpha is a smoothing parameter

# Initialize GridSearchCV
grid_search = GridSearchCV(nb_model, param_grid, cv=5, scoring='accuracy')

# Perform hyperparameter tuning on the training data
grid_search.fit(X_train, y_train)

# Get the best model and its parameters
best_nb_model = grid_search.best_estimator_
best_params = grid_search.best_params_

print(f"Best Parameters: {best_params}")

# Predict on the test set using the best model
y_pred = best_nb_model.predict(X_test)

# Evaluate the best model
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy with Best Model: {accuracy}")
print("Classification Report with Best Model:")
print(report)
```
